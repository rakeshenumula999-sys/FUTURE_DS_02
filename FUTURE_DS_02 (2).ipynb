# Customer Retention & Churn Analysis

## Objective
The objective of this project is to analyze customer churn patterns in a subscription-based business.

This analysis identifies:
- Churn rate
- Retention performance
- Revenue loss
- Key churn drivers
- Actionable recommendations

Churn refers to customers who discontinue their subscription service.
Retention refers to customers who continue using the service.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("customer_churn_data.csv")
df.head()
```

## Data Overview

The dataset contains customer-level subscription data including:
- Subscription Type
- Monthly Charges
- Tenure (Months)
- Churn Status

The dataset was checked for missing values and duplicates before analysis.

```python
df.isnull().sum()
df.duplicated().sum()
```

```python
total_customers = len(df)
churned_customers = len(df[df['Churn'] == 'Yes'])
retained_customers = len(df[df['Churn'] == 'No'])

churn_rate = (churned_customers / total_customers) * 100
retention_rate = 100 - churn_rate

print("Total Customers:", total_customers)
print("Churned Customers:", churned_customers)
print("Retention Rate:", retention_rate)
print("Churn Rate:", churn_rate)
```

## KPI Summary

- Total Customers: 20
- Churned Customers: 8
- Retained Customers: 12
- Churn Rate: 40%
- Retention Rate: 60%

The company currently has a churn rate of 40%, meaning 4 out of every 10 customers discontinue their subscription.
This indicates a high churn risk for the business.

```python
sns.countplot(x='Churn', data=df)
plt.title("Churn Distribution")
plt.show()
```

```python
sns.countplot(x='Subscription_Type', hue='Churn', data=df)
plt.title("Subscription Type vs Churn")
plt.show()
```

```python
sns.boxplot(x='Churn', y='Tenure_Months', data=df)
plt.title("Tenure vs Churn")
plt.show()
```

```python
revenue_loss = df[df['Churn'] == 'Yes']['Monthly_Charges'].sum()
print("Revenue Lost Due to Churn:", revenue_loss)
```

## Key Insights

1. Customers with Basic subscription plans show higher churn compared to Premium plans.

2. Customers with lower tenure months are more likely to churn, indicating onboarding issues.

3. Churn results in significant revenue loss.

4. Early-stage customers represent the highest churn risk group.

## Business Recommendations

1. Improve onboarding during the first 3 months.
2. Encourage customers to upgrade to higher-value plans.
3. Introduce loyalty rewards for long-term subscribers.
4. Launch targeted retention campaigns for high-risk customers.

## Conclusion

This analysis identified key churn drivers including subscription type and tenure.

Reducing churn by even 10% can significantly improve long-term profitability.

The company should focus on early engagement and retention strategies to improve customer lifetime value.
